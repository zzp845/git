```python
text = 'А судьи кто? – За древностию лет К свободной жизни их вражда непримирима, Сужденья черпают из забытых газет Времен Очаковских и покоренья Крыма; Всегда готовые к журьбе, Поют всё песнь одну и ту же, Не замечая об себе: Что старее, то хуже. Где, укажите нам, отечества отцы, Которых мы должны принять за образцы? Не эти ли, грабительством богаты? Защиту от суда в друзьях нашли, в родстве, Великолепные соорудя палаты, Где разливаются в пирах и мотовстве И где не воскресят клиенты‑иностранцы Прошедшего житья подлейшие черты. Да и кому в Москве не зажимали рты Обеды, ужины и танцы?'
nlp = spacy.load("ru_core_news_sm")
doc = nlp(text)

spacy_text1 = nlp(text)
spacy_text1
```

### 1. Токенизация

Токенизация — это процесс разбиения текста на отдельные элементы: слова, знаки препинания и специальные символы. Она необходима для дальнейшего анализа каждого компонента текста по отдельности.

Результатом является список токенов, полученных из исходного текста.


```python
print("Токенизация:")
tokens = [token.text for token in doc if not token.is_space]
print(tokens)
```

### 2. Частеречная разметка

На этом этапе каждое слово получает грамматическую метку — часть речи (например, существительное, глагол и т.д.). Это важно для синтаксического анализа и последующих шагов в NLP.

Для каждого токена указана его часть речи и краткое пояснение её значения.


```python
print("Части речи с пояснениями:")
for token in doc:
    if not token.is_punct and not token.is_space:
        pos = token.pos_
        explanation = spacy.explain(pos)
        print(f"{token.text:20} — {pos:5} — {explanation}")
```

### 3. Лемматизация

На данном этапе производится приведение слов к их начальной (словарной) форме — лемме. Это необходимо для нормализации текста и дальнейшей корректной обработки (например, при анализе частоты слов или классификации).

В таблице ниже приведено соответствие исходных слов и их лемм.


```python
print("Лемматизация:")
for token in doc:
    if not token.is_punct and not token.is_space:
        print(f"{token.text} → {token.lemma_}")
```

### 4. Выделение именованных сущностей

Именованные сущности — это значимые элементы текста, такие как имена людей, названия организаций, даты и географические объекты. Их автоматическое извлечение важно для структурирования информации и последующего анализа.

Ниже приведён список сущностей, распознанных в тексте, с пояснением типа каждой.


```python
print("Именованные сущности:")
for ent in doc.ents:
    explanation = spacy.explain(ent.label_)
    print(f"{ent.text:40} — {ent.label_:10} — {explanation}")
```
